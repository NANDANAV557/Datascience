```python
row = 5
for i in range(1, row + 1):
  print("*" * i)
```

```python
row = 5
for i in range(row, 0, -1):
  print("*" *i)
```

```python
n = 5
for i in range(n):
  for j in range(n):
    print("*",end=" ")
  print()
```

```python
rows = 5

for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
rows = 5

for i in range(rows):
    print(" " * (rows - i - 1) + "* " * (i + 1))
```

```python
rows = 5

for i in range(rows, 0, -1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
rows = 5

for i in range(1, rows + 1):
    ch = 'A'
    for j in range(i):
        print(ch, end=" ")
        ch = chr(ord(ch) + 1)
    print()
```
